```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
from PIL import Image
```

```python
def view_image_and_mask(image_path, mask_path):
    """
    View image and mask
    """
    if not os.path.exists(image_path):
        print("Image path does not exist")
        return None
    image = cv2.imread(image_path, 1)
    mask = cv2.imread(mask_path)
    fig, ax = plt.subplots(1,2, figsize=(10, 10))
    ax[0].imshow(image)
    ax[1].imshow(mask[:,:,0])
    plt.show()
```

```python
image_path = os.path.join(os.getcwd(), "../data", "selected_data","images", "M-33-7-A-d-2-3_patch_0_8.tif")
mask_path = os.path.join(os.getcwd(), "../data", "selected_data","masks", "M-33-7-A-d-2-3_patch_0_8.tif")
view_image_and_mask(image_path, mask_path)
```
